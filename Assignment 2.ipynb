###### Rearranging Authors names so that first name is displayed first before the second name to enable earsier searchability and readability for the data.
###### Setting values that hold no data to unknown.
###### Saving reordered data from books.csv to a dictionary.
###### Formatting date so that it reads yyyymmdd insead of the excel days since epoch date to allow for easier readability of the data.
###### Saving reordered data from bookloans.csv to a dictionary.
###### members.csv data is read and reformated so that name is saved as full name, ordered first name and then last name, as one value to enable better readability and searchability of the data.

```python
# Importing relevant libraries
import csv
import json
import datetime
```

```python
#set any missing values to unknown, rearrange authors names so that their first name is displayed first, save all data to dict

def NameFormatting(name):
    # split the name and reorder the name where name has a comma
    if (',' in name) and ('-' not in name):
        split_id = name.split(', ')
        first = split_id[0]
        second = split_id[1]
        return second + ' ' + first
    # split the name and reorder the name where name has comma and hyphen
    elif (',' in name) and ('-' in name):
        split_hyphen = name.split('-')
        prehyphen = split_hyphen[0]
        posthyphen = split_hyphen[1]
        split_comma = prehyphen.split(', ')
        second0 = split_comma[0]
        first0 = split_comma[1]
        return first0 + second0 + posthyphen
    # rename values that hold no data to unknown
    elif len(name) == 0:
        name = 'Unknown'
        return name
    # if none of the above are required no edits are made
    elif len(name) > 0:
        return name

with open('books.csv', mode = 'r', encoding = 'utf-8-sig') as file:
    csvFile = csv.reader(file)
    # set up empty dict
    books_dict = {}
    # set up empty list
    books_list = []
    # set up row counter
    rowCount = 0
    for row in csvFile:
        # capture row names for dict
        if rowCount == 0:
            name0 = row[0]
            name1 = row[1]
            name2 = row[2]
            name3 = row[3]
            name4 = row[4]
            name5 = row[5]
            rowCount += 1
        else:
            books_list += [row[0]]
            # program is told how to display data in the dictionary
            books_dict[row[0]] = {name1: NameFormatting(row[1]), name2: NameFormatting(row[2]), name3: row[3], name4: row[4], name5: row[5], "Available": "Yes", "Reserved": "No", "ReservedID": ()} 
            rowCount ++ 1

# check functionality
print(books_dict)
    
```

```python
# read bookloans.csv data file and save to dict
def DateFormatting(refdate, value):
# function to  convert value representing days passed since epoch to standard yyyymmdd
    # refdate represents epoch
    refDate = "1900-01-01"
    # tell program format of refdate
    formatref = datetime.datetime.strptime(refdate, "%Y-%m-%d")
    # calculation to create new date from epoch, 
    newdate = (formatref + datetime.timedelta(days=int(value))).strftime("%Y-%m-%d")
    return newdate

with open('bookloans.csv', mode = 'r', encoding = 'utf-8-sig') as file:
    csvFile = csv.reader(file)
    # set up row counter
    rowCount = 0
    # set up empty dict
    bookloans_dict = {}
    for row in csvFile:
        # only output loan date if book exists in books_dict
        if row[0] in books_list:
            rowCount += 1
            if row[3] != 0:
                dateOfReturn = DateFormatting('1900-01-01', row[3])
                lengthBorrowed = int(row[3]) - int(row[2])
            else:
                dateOfReturn = "NaN"
                lengthBorrowed = "NaN"
                books_dict[row[0]]["Available"] = "No"
            # program is told how to display data in the dictionary
            bookloans_dict[rowCount] = {'Book ID': row[0], 'Member ID': row[1], 'Date Of Loan': DateFormatting('1900-01-01', row[2]), 'Date Of Return': dateOfReturn, 'Length Borrowed': lengthBorrowed}

# check functionality
print(bookloans_dict)
```

```python
# members csv is opened for reading
with open('members.csv', mode = 'r', encoding = 'utf-8-sig') as file:
    # csv reader defined as csv file
    csvFile = csv.reader(file)
    # row count set to 0
    rowCount = 0
    # empty dictionary created
    members_dict = {}
    for row in csvFile:
        if rowCount == 0:
            # row count increased by 1 to allow for cycling through data
            rowCount += 1
            # rows defined
            name0 = row[0]
            name1 = row[1]
            name2 = row[2]
            name3 = row[3]
            name4 = row[4]
            name5 = row[5]
        else:
            # program is told how to display data in the dictionary
            members_dict[row[0]] = {name1: row[1], name2: row[2], "Name": (row[1]+ " " +row[2]), name3: row[3], name4: row[4], name5: row[5]}
            rowCount += 1

# check functionality
print(members_dict)
            
```

```python
# using w as mode JSON files are created for each dictionary. The dictionaries are written to the JSON files.
with open('books.json', 'w') as s:
    json.dump(books_dict, s)
    print('books.json saved')
with open('loans.json', 'w') as s:
    json.dump(bookloans_dict, s)
    print('loans.json saved')
with open('members.json', 'w') as s:
    json.dump(members_dict, s)
    print('members.json saved')
```

```python
# defining loadJSON as a function to open JSON files
def loadJSON(file):
    f = open(file)
    file = json.load(f)
    f.close()
    return file

# opening JSON files
books = loadJSON("books.json")
print("books.json loaded")
loans = loadJSON("loans.json")
print("loans.json loaded")
members = loadJSON("members.json")
print("members.json loaded")
```

```python
# Book created as a class
class Book(object):
    """ Pre-Conditions - Assumes that there is a dictionary called Books acsessible and that within the Books dictionary there is a key called Title
        Post-Conditions - Returns all information regarding a specific Book from the Book dictionary, using the Book Title to search. If Book is found in the Library all information is returned. If the book is not found in the library an error message is printed."""
    # tells class that scan is a method that should be inherited as a subclass
    @classmethod
    def scan(cls, title):
        ID = [key for key in books if books[key]["Title"]== title]
        # if book ID is not empty the id will be returned
        if len(ID) >= 1:
            return ID
        # if book ID is empty it is not in library and the program will print a message
        elif len(ID) == 0:
            return print(title, "was not found in the library.")

#functionality check
Lookedupbook = Book.scan(input("Book Search: "))
print(books[Lookedupbook[0]])
```

```python
# Member created as a class
class Member(object):
    """Pre-Conditions - Program assumes that there is a ductionary called members and within that dictionary there is a key called Name
        Post-Conditions - Returns all member information in the member dictionary, using the member's name to search. If member is not found in the dictionary a message is printed for the librarian"""
    # tells class that scan is a method that should be inherited as a subclass
    @classmethod
    def scan(cls, name):
        ID = [key for key in members if members[key]["Name"]== name]
        # if member ID is not empty the id will be returned
        if len(ID) >= 1:
            return ID
        # if member ID is empty the person does not have an account and the program will print a message
        elif len(ID) == 0:
            return print(name, "does not have an account with the library.")

# functionality check
Lookedupmember = Member.scan(input("Member Search: "))
print(members[Lookedupmember[0]])
```
